```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

def laplacian_transform(image):
    # Applying the Laplacian operator to the image
    laplacian = cv2.Laplacian(image, cv2.CV_64F)
    return laplacian

def main(directory_path):
    # Get the list of image file names in the specified directory
    image_files = [f for f in os.listdir(directory_path) if f.endswith(".jpg") or f.endswith(".png")]

    for image_file in image_files:
        # Read the grayscale image
        image_path = os.path.join(directory_path, image_file)
        grayscale_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

        # Apply the Laplacian transform
        laplacian_output = laplacian_transform(grayscale_image)

        # Plot the original and transformed images side by side
        plt.figure(figsize=(10, 5))
        plt.subplot(1, 2, 1)
        plt.imshow(grayscale_image, cmap="gray")
        plt.title("Original Grayscale Image")
        plt.axis("off")

        plt.subplot(1, 2, 2)
        plt.imshow(laplacian_output, cmap="gray")
        plt.title("Laplacian Transform Output")
        plt.axis("off")

        plt.show()

if __name__ == "__main__":
    # Replace 'your_directory_path' with the actual path to the directory containing your grayscale images
    directory_path = "/content/input-images"
    main(directory_path)
```
